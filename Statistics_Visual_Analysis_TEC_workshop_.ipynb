##2. Clean the dataset by removing the empty column

```python
import pandas as pd

# Load the dataset
data_frame = pd.read_csv('titanic.csv')

# Remove rows where any cell in the specified columns empty
data_frame_cleaned = data_frame.dropna(subset=['survived', 'gender', 'age', 'fare', 'num_siblings', 'num_parents'])

# Save the cleaned data to a new CSV file
data_frame_cleaned.to_csv('titanic_cleaned.csv', index=False)
```

###4. What is distribution of male and female passengers in the dataset [**Bar plot**]?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'titanic_cleaned.csv'
data = pd.read_csv(file_path)

# Initialize counts
num_male = 0
num_female = 0

# Counting the number of patients with depression (1) and without depression (0)
for value in data['gender']:
    if value == 'male':
        num_male += 1
    elif value == 'female':
        num_female += 1

# Plotting the results with horizontal bars
labels = ['Male', 'Female']
counts = [num_male, num_female]

plt.figure(figsize=(6, 2))
plt.barh(labels, counts, color=['red', 'green'], height = 0.5)  # Horizontal bar plot
plt.xlabel('Number of Passengers')
plt.ylabel('Passengers gender')
plt.title('Distribution of passengers')
plt.show()
```

##7. How is the distribution of passengers' ages divided into quartiles [**Box plot**]?

##9. What is the percentage distribution of unique values in the num_siblings column of the Titanic dataset?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'titanic_cleaned.csv'
data = pd.read_csv(file_path)

# Count the occurrences of unique values in the 'num_siblings' column
siblings_counts = data['num_siblings'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(siblings_counts, labels=siblings_counts.index, autopct='%1.1f%%', colors=plt.get_cmap('tab10').colors)
plt.title('Percentage Distribution of Unique Values in Number of Siblings')
plt.show()
```

What is the percentage of passengers who survived?

```python
# prompt: What is the percentage of passengers who survived?

# Load the dataset
file_path = 'titanic_cleaned.csv'
data = pd.read_csv(file_path)

# Calculate the percentage of passengers who survived
total_passengers = len(data)
survived_passengers = len(data[data['survived'] == 1])
percentage_survived = (survived_passengers / total_passengers) * 100

# Print the result
print(f"The percentage of passengers who survived is: {percentage_survived:.2f}%")
```

What is the percentage of passengers who did not survive?

```python
# prompt: What is the percentage of passengers who did not survive?

# Calculate the percentage of passengers who did not survive
total_passengers = len(data)
not_survived_passengers = len(data[data['survived'] == 0])
percentage_not_survived = (not_survived_passengers / total_passengers) * 100

# Print the result
print(f"The percentage of passengers who did not survive is: {percentage_not_survived:.2f}%")
```

What is the distribution of passenger classes (1st, 2nd, 3rd)?

```python
# prompt: What is the distribution of passenger classes (1st, 2nd, 3rd)?

# Load the dataset
file_path = 'titanic_cleaned.csv'
data = pd.read_csv(file_path)

# Count the occurrences of each passenger class
class_counts = data['fare'].value_counts()

# Print the distribution
print("Distribution of Passenger Classes:")
for fare, count in class_counts.items():
  print(f"Class {fare}: {count} passengers")

# Plotting the distribution as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(class_counts.index, class_counts.values)
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Passenger Classes')
plt.xticks(class_counts.index)  # Set x-axis ticks to be the class numbers
plt.show()
```
